```python
country_names = final_df.Country.value_counts().index
```

```python
names =  country_names[:3]
```

```python
country_val = final_df.Country.value_counts().values
```

```python
#pie chart - Top 3 countries that uses Zomato
plt.pie(country_val[:3],labels =names,autopct = '%1.2f%%');
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0:'RatingCount'})
```

```python
ratings.head()
```

'''OBSERVATIONS
1.when rating is between 4.5 to 4.9 - Excellent
2. when rating is between 4.0 to 3.4 - Very Good
3. when rating is between 3.5 to 3.9 - good
4. when rating is between 2.5 to 3.4 - average
5. when rating is betweem 1.0 to 2.4  - poor
6. if rating is 0 then Not rated'''

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="RatingCount",hue = 'Rating color',data = ratings,pallete=['white','red','orange','yellow','green','green'])
```
